```python
import pandas as pd
import numpy as np
import nltk
import warnings

%matplotlib inline    

warnings.filterwarnings('ignore')
```

```python

dataset = pd.read_csv('/content/Youtube04-Eminem.csv')
dataset
```

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def tokenize(comment):
    return word_tokenize(comment)
```

```python
dataset['tokens'] = dataset['CONTENT'].apply(tokenize)
```

```python
print('Tokens for first comment:')
print(dataset['tokens'][0])
```

```python
from nltk.stem import SnowballStemmer
def stem(comment):
    stemmer = SnowballStemmer('english')
    return [stemmer.stem(word) for word in word_tokenize(comment)]

# Apply the stemming function to each comment in the dataset
dataset['stemmed'] = dataset['CONTENT'].apply(stem)

# Print the stemmed words for the first comment in the dataset
print('Stemmed words for Third comment:')
print(dataset['stemmed'][2])
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
def lemmatize(comment):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word, pos='v') for word in word_tokenize(comment)]

# Apply the lemmatization function to each comment in the dataset
dataset['lemmatized'] = dataset['CONTENT'].apply(lemmatize)

# Print the lemmatized words for the first comment in the dataset
print('Lemmatized words for Third comment:')
print(dataset['lemmatized'][2])
```

```python
new_dataset = dataset[['CONTENT', 'CLASS']]
new_dataset
```

```python
import re

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
new_dataset['CONTENT'] = np.vectorize(remove_pattern)(new_dataset['CONTENT'], "@[\w]*")
```

```python
new_dataset['CONTENT'] = new_dataset['CONTENT'].str.replace("[^a-zA-Z#]", " ")
new_dataset.head()
```

```python
new_dataset['CONTENT'] = new_dataset['CONTENT'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))
new_dataset.head()
```

```python
new_dataset['CONTENT'] = new_dataset['CONTENT'].apply(lambda x: x.lower())
new_dataset
```

```python
nltk.download('all')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
from textblob import TextBlob

def get_sentiment(comment):
    blob = TextBlob(comment)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'


dataset['sentiment'] = dataset['CONTENT'].apply(get_sentiment)

sentiment_counts=print(dataset['sentiment'].value_counts())
```

```python
print('\nSample comments by sentiment:')
for sentiment in ['positive', 'negative', 'neutral']:
    sample = dataset[dataset['sentiment'] == sentiment]['CONTENT'].sample(5, random_state=1)
    print(f'{sentiment.capitalize()} comments:\n{sample}\n')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1420)
```

```python
X = cv.fit_transform(new_dataset['CONTENT']).toarray()
y = dataset.iloc[:, -1].values
```

```python
import pickle
bow_path = 'c1_BoW_Sentiment_Model.pkl'
pickle.dump(cv, open(bow_path, "wb"))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
import joblib
joblib.dump(classifier, 'c2_Classifier_Sentiment_Model') 
     
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
matrx = confusion_matrix(y_test, y_pred)
print(matrx)

accuracy_score(y_test, y_pred)
```
